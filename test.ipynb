```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
```

```python
url = "https://web.dev/howbrowserswork/"
stop_html_elements = ["script", "style", "nav", "footer", "img"]
```

```python
page = requests.get(url)
```

```python
html = page.content
```

```python
soup = BeautifulSoup(html, features="html.parser")
```

```python
all = soup.find_all("meta")
all
```

```python
soup(all[0]).decompose()
soup
```
