```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
f = open("output.html", "r")
html = f.read()
f.close()
```

```python
soup = BeautifulSoup(html, features="html.parser")
```

```python
stop_html_elements = ["head", "script", "style",
                                   "footer", "img", "button",
                                   "h1", "h2", "h3",
                                   "header", "web-header", "meta",
                                   "link", "nav"]


for element in soup.findAll(stop_html_elements):
    element.decompose()
```

```python
text_elements = soup.findAll(["p", "ol"])
```

```python
whole_text = ""
for text_element in text_elements:
    whole_text += text_element.text + "\n"
```

```python
with open("output_text.txt", "w") as f:
    f.write(whole_text)
```
